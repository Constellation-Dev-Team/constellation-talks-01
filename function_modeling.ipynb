```python
import pandas as pd
import numpy as np
from models.perceptron import Perceptron
from models.single_layer_perceptron import SingleLayerPerceptron
```

```python
def gen(X):
    return X[0] * 2 + 3

sample_size = 2000
X = np.random.uniform(0,100, (sample_size, 1))
y = [gen(i) for i in X]

neuron = Perceptron(1)
neuron.learning_rate = 2e-4
print(neuron.train(X,y))
neuron.predict_single([21]), gen([21])
```

```python
n_features = 5
number_to_generate = 1000

def gen(numbers):
    return [x[0]*2 + 2*x[1] + x[2]*x[3] - x[4] + 3 for x in numbers]
X = np.random.uniform(0, 100, size=(number_to_generate, n_features))
y = gen(X)

slp = SingleLayerPerceptron(n_features, layer_size=4)
slp.learning_rate = 1e-5
print(slp.train(X,y))
print(slp.predict_single([1, 1, 1, 1, 1]) , gen([[1, 1, 1, 1, 1]]))
pd.Series(slp.saved_errors[:]).plot()
```

```python
print(slp.predict_single([1, 1, 1, 1, 1]) , gen([[1, 1, 1, 1, 1]]))
pd.Series(slp.saved_errors[:]).plot()
```

```python
from matplotlib import pyplot as plt
plt.yscale('symlog')
pd.Series(slp.saved_errors[:100]).plot()
```

```python
slp = SingleLayerPerceptron(n_features)
slp.learning_rate = 1e-2
print(slp.train(X,y))
print(slp.predict_single([1, 1, 1, 1, 1]) , gen([[1, 1, 1, 1, 1]]))
plt.yscale('symlog')
pd.Series(slp.saved_errors[:100]).plot()
```

```python
y_pred = slp.predict(X)

pd.DataFrame([y, y_pred[:, 0]]).transpose().plot()
```
