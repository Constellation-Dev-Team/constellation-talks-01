```python
n_features = 5
number_to_generate = 100

def gen(numbers):
    return [x[0] + 2*x[1] + 3*x[2] + x[3] + -1*x[4] + 3 for x in numbers]
X = np.random.randint(0, 100, size=(number_to_generate, n_features))
y = gen(X)

slp = SingleLayerPerceptron(n_features)
slp.learning_rate = 0.02
print(slp.train(X,y))
print(slp.predict_single([1, 1, 1, 1, 1]) , gen([[1, 1, 1, 1, 1]]))
pd.Series(slp.saved_errors[:]).plot()
```

```python
from matplotlib import pyplot as plt
plt.yscale('symlog')
pd.Series(slp.saved_errors[:100]).plot()
```
